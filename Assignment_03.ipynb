### 1. What is the determinant of a matrix?

<font color = "darkred">Point(s) = 1</font>

The determinant of a matrix is a function within linear algebra that converts a matrix $A$ into a scalar. However, it is important to note that the determinant is only valid for square matrices with n x n dimensions. The determinant of a 2x2 matrix, $A$, is the following: 

$$ det(\textbf{A}) = |\textbf{A}| =  ad-bc $$

In addition to this, the determinant has the following important properties:
1. Determine if a matrix is invertible i.e. if det (A) $\neq$ 0, then it is invertible.
2. A determinant can be used to assess the linear independence of vectors.
3. A determinant can be used to compute eigenvalues.
4. A determinant can be used to solve systems of linear equations.

### 2. What is an eigenvector?

<font color = "darkred">Point(s) = 1</font>

An eigenvector is a vector that satisfies the eigenvalue equation for a particular eigenvalue (Moore & Siegel 333). Specifically, the eigenvalue equation is: \begin{equation}
\textbf{A}\vec{v} = \lambda \vec{v}
\end{equation}

Further, one can interpret the above equation by stating that eigenvectors are “vectors that, when multiplied by $A$, end up as scalar multiples of themselves” (Moore & Siegel 335). Eigenvectors have three main use cases in political science:
1. Matrix decomposition and principal components analysis
2. Assessing degrees of centrality for an actor in a network
3. Interpreting x as the state of the system and A as the transition matrix between states (Moore & Siegel 335).



### 3. How do you determine if two vectors are orthogonal?

<font color = "darkred">Point(s) = 1</font>

Two vectors are orthogonal when the two vectors are perpendicular to each other. Further, one can determine if \vec{a} and \vec{b} are orthogonal because the cosine of the angle between \vec{a} and \vec{b} will equal 0, or specifically: $$\cos{\theta} = 0$$

Additionally, the dot product of two orthogonal vectors will equal zero: $$ \vec{a} \cdot \vec{b} =  0 $$

### 5. What does the following equation tell us and why does it matter?

$$ ||\vec{a} + \vec{b} || \le || \vec{a} || + || \vec{b} || $$

<font color = "darkred">Point(s) = 1</font>

The above equation refers to the triangle inequality. The triangle inequality states that the length of a sum of vectors $||\vec{a} + \vec{b} ||$ can be equal to, but cannot exceed, the length of a sum of their individual vectors $|| \vec{a} || + || \vec{b} ||$ (Moore & Siegel, 279).

### 6. What is the inverse of a matrix?

<font color = "darkred">Point(s) = 1</font>

A matrix is invertible if a second square matrix, $B$, when multiplied with $A$, produces the identity matrix. It is important to note that only square matrices can be inverted. Additionally, being able to invert a matrix indicates that a system of linear equations has a solution. The following properties are true when inverting matrices:

$AB = BA = I$

Or specifically:

$A{A}^{-1} = {A}^{-1}A = I$

### 7. What does it mean that two vectors are linearly independent?

<font color = "darkred">Point(s) = 1</font>

Two vectors are linearly independent when no vector in a set is a linear combination of other vectors in this set (Moore & Siegel 307). Specifically, two vectors are linearly independent when their coefficients equal zero. In the case below, coefficients $a_1 \dots a_n = 0$

The below set of vectors are linearly independent:

$$ a_1 v_1 + a_2 v_2 + \dots + a_n b_n = 0 $$

In addition to this, linear independence produces the following insights:
1. Linear independence tells when systems of equations have solutions
2. Linear independence tells the size of a vector space
	1. A set of linearly independent vectors that span a vector space is referred to as a basis.
	2. The dimension of a vector space equals the number of vectors in its basis (Moore & Siegel 	308)

```python
# Calculate the betas.
B = la.inv(U.T.dot(U)).dot(U.T.dot(v))
B = B.round(3)
```

```python
print("Linear coefficients are: " + str(B))
```

```python
# Calculate the model fit R^2
y_hat = U.dot(B)
y_hat_center = y_hat - y_hat.mean()
y_center = v - v.mean()
r_squared = y_hat_center.dot(y_hat_center)/y_center.dot(y_center)
```
